```python
import pandas as pd

years = [2019, 2018, 2017]
grade = 3
subjects = ['Math', 'Swedish', 'Swedish_as_a_second_language']

df_dict = {}
for i, s in enumerate(subjects):
    for year in years:
        title = f'{year}_grade-{grade}_{s}'
        df_dict[title] = pd.read_excel(f'data/exp_ap{grade}_{year}.xlsx',
                                   sheet_name=i, 
                                   header=7
                                  )
```

```python
years = [2019, 2018, 2017]
grade = 6
subjects = ['English', 'Math', 'Swedish', 'Swedish_as_a_second_language']

for i, s in enumerate(subjects):
    for year in years:
        title = f'{year}_grade-{grade}_{s}'
        df_dict[title] = pd.read_excel(f'data/exp_ap{grade}_2013_{year}.xlsx',
                                   sheet_name=i, 
                                   header=7
                                  )
```

```python
df_dict.keys()
```

```python
school_codes = set()
for tmp_codes in [df_dict[k]['Skol-enhetskod'].unique() for k in df_dict.keys()]:
    for code in tmp_codes:
        school_codes.add(code)
```

```python
len(school_codes)
```

```python
school_info = pd.read_excel(f'data/Skolenhetsregistret_20220602.xlsx',
                                   sheet_name='Grundskola'
                                  )

school_info.head()
```

```python
relevant_school_info = school_info.query('SKOLENHETSKOD in @school_codes')
relevant_school_info.head()
```
