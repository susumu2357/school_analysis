# Grade 6

Focus on the sixth grade because the statistics gives absolute score for each subject.

Use the "Genomsnittlig provbetygspoäng" column.
The description of the columns:

>The average test score shows an average of the students' test scores converted into points. The average test score is calculated for pupils who have obtained test grades A-F in the respective section or subject tests. The grade F gives 0 points, E gives 10 points, D gives 12.5, C gives 15, B gives 17.5 and A gives 20 points. The highest possible test mark is 20 points and the lowest is 0 points.

- `columns[-9]` corresponds to the number of students.
-  `columns[-3]` corresponds to the score.

These columns have the same name "Totalt" due to the excel column structure.

```python
import numpy as np

years = [2019, 2018, 2017]
grade = 6
subjects = ['English', 'Math', 'Swedish', 'Swedish_as_a_second_language']

info_cols = ['SKOLENHETSKOD', 'HUVUDMANNATYP', 'KOMMUNNAMN', 'SKOLENHETENS NAMN', 'BESÖKSADRESS', 'BESÖKSPOSTNR', 'BESÖKSPOSTORT', 'WEBB', 'FKLASS']

summary_dict = {}

for i, s in enumerate(subjects):
    summary_dict[s] = []
    for year in years:
        title = f'{year}_grade-{grade}_{s}'
        cols = ['Skol-enhetskod', 'Delprov'] + [df_dict[title].columns[-9]] + [df_dict[title].columns[-3]]
        renamed_cols = ['school_code', 'subtest', 'num_students', 'score']
        tmp_df = df_dict[title][cols].rename(columns={
            k:v for k,v in zip(cols, renamed_cols)
        })
        tmp_df['score'] = tmp_df['score'].replace('.', np.nan).replace('..', np.nan).astype(float)
        tmp_df = tmp_df.pivot(index='school_code', columns='subtest', values='score')
        tmp_df['avg_score'] = tmp_df.mean(axis=1, skipna=True)
        tmp_df = tmp_df.merge(relevant_school_info[info_cols], left_on='school_code', right_on='SKOLENHETSKOD', how='inner')
        tmp_df['year'] = year
        summary_dict[s].append(tmp_df)
    summary_dict[s] = pd.concat(summary_dict[s]).reset_index(drop=True)
        
```

```python
summary_dict['Swedish'].head()
```

```python
tmp_list = []

for i, s in enumerate(subjects):
    tmp_df = (
        summary_dict[s]
        .groupby('SKOLENHETSKOD')['avg_score']
        .apply(lambda x:x.mean(skipna=True))
        .reset_index()
        .rename(columns={'avg_score': f'{s}_avg_score'})
    )
    tmp_list.append(tmp_df)

avg_df = tmp_list[0]
for elm in tmp_list[1:]:
    avg_df = avg_df.merge(elm, on='SKOLENHETSKOD', how='outer')
avg_df['grand_avg_score'] = avg_df[avg_df.columns[avg_df.columns.str.endswith('score')]].mean(axis=1, skipna=True)
avg_df = avg_df.merge(relevant_school_info[info_cols], on='SKOLENHETSKOD')
avg_df = avg_df.dropna(subset=['grand_avg_score']).reset_index(drop=True)
avg_df.sort_values('grand_avg_score', ascending=False)
```

```python
import json
from tqdm import tqdm
import requests
import os

API_KEY = os.environ['API_KEY']
lats = []
lons = []
ratings = []

for i, row in tqdm(avg_df.iterrows(), total=len(avg_df)):
    s = ' '.join(row[['SKOLENHETENS NAMN', 'BESÖKSPOSTNR', 'BESÖKSADRESS', 'BESÖKSPOSTORT']].astype(str).values) + ' Sweden'
    address = s.replace(' ', '%20')
    url = f'https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={address}&inputtype=textquery&fields=rating%2Cgeometry&key={API_KEY}'

    payload={}
    headers = {}

    response = requests.request("GET", url, headers=headers, data=payload)
    res = json.loads(response.text)
        
    if len(res['candidates']) == 0:
        print(i)
        url = f'https://maps.googleapis.com/maps/api/geocode/json?address={address}&key={API_KEY}'
        payload={}
        headers = {}
        response = requests.request("GET", url, headers=headers, data=payload)
        res = json.loads(response.text)
        try:
            loc = res['results'][0]['geometry']['location']
        except:
            loc = {'lat': np.nan, 'lng':np.nan}
        
        lats.append(loc['lat'])
        lons.append(loc['lng'])
        ratings.append(np.nan)
    else:   
        loc = res['candidates'][0]['geometry']['location']
        lats.append(loc['lat'])
        lons.append(loc['lng'])
        try:
            ratings.append(res['candidates'][0]['rating'])
        except:
            ratings.append(np.nan)
```

```python
avg_df['lon'] = lons
avg_df['lat'] = lats
avg_df['Google_rating'] = ratings

avg_df.head()
```

```python
avg_df['Google_rating'] = avg_df['Google_rating'].apply(lambda x: np.nan if x==0 else x)
```
